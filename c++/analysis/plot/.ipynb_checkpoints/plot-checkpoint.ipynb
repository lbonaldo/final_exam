```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#I create the file

n_keys = [50, 100, 200, 500, 1000, 2000, 3000, 50000]
stat_tree = ['unbal','bal','std']
type_tree = ['ID','DD','SD','SS']

filelist=[]

for n_key in n_keys:
    for stat in stat_tree:
        for ttype in type_tree:
            filelist.append("%s_%s_mean_%s" % (ttype,stat,n_key))
            
print(filelist)
```

```python
#create the plot

#intdim = np.int_(dim)
i=0
for i,file in enumerate(filelist):
    if(i<12):
        plt.figure(figsize=(15,13))
        dim, time = np.loadtxt("../data/mean/"+file+".txt", delimiter = ' ', unpack = True)
        plt.plot(dim, time, marker = 'o')
        plt.title(file+' - look-up time', fontsize=24)
        plt.xlabel('dim', fontsize=22)
        #plt.xticks(dim,intdim)
        plt.ylabel('time [us]', fontsize=22)
        plt.savefig("./plots/%s.png" % file, bbox_inches='tight')
```
