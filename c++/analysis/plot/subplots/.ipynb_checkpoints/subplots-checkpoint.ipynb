```python
%pylab inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
#I create the file

n_keys = [50, 100, 200, 500, 1000, 2000, 3000, 50000]
stat_tree = ['unbal','bal','std']
type_tree = ['ID','DD','SD','SS']

filelist=[]

for n_key in n_keys:
    for ttype in type_tree:
        for stat in stat_tree:
            filelist.append("%s_%s_mean_%s" % (ttype,stat,n_key))
            
print(filelist)
```

```python
dim0, time0, err0 = np.loadtxt("../../data/mean/"+filelist[0]+".txt", delimiter = ' ', unpack = True)
dim1, time1, err1 = np.loadtxt("../../data/mean/"+filelist[1]+".txt", delimiter = ' ', unpack = True)
dim2, time2, err2 = np.loadtxt("../../data/mean/"+filelist[2]+".txt", delimiter = ' ', unpack = True)

plt.figure(figsize=(15,13))
p0 = plt.errorbar(dim0, time0, yerr=err0, marker = 'o')
p1 = plt.errorbar(dim1, time1, yerr=err1, marker = 'o')
p2 = plt.errorbar(dim2, time2, yerr=err2, marker = 'o')
plt.title('ID_50 look-up time', fontsize=24)
plt.xlabel('dim', fontsize=22)
plt.xscale('linear')       
#plt.xticks(dim,intdim)
plt.ylabel('time [us]', fontsize=22)
plt.grid(True, which='major', linestyle='--')  
plt.legend((p0[0], p1[0], p2[0]), ('unbalanced', 'balanced', 'std::map'), fontsize=20)
plt.savefig("./plots/ID_50 look-up time.png", bbox_inches='tight')
```
