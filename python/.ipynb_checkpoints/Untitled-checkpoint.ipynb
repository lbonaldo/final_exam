```python
import collections
import datetime

class PostcardList():
    
    def __init__(self): #, _postcards, _date, _from, _to):
        
        self.file_name = None
        self._postcards = []
        self._date = {}
        self._from = {}
        self._to = {}
        
    def readFile(self, from_file_name):
        self.file_name = from_file_name
        with open(from_file_name, 'r') as file:
            for line in file:
                self._postcards.append(line)
                
            self.parsePostcards()
            
    def writeFile(self, to_file_name):
        with open(to_file_name, 'w') as file:
            for line in self._postcards:
                file.write(line)
                
    def updateFile(self, new_postcards):
        with open(self.file_name, 'a+') as file:
            for line in new_postcards:
                file.write(line)
                
    def updateList(self, from_file_name):
        with open(from_file_name, 'r') as file:
            for line in from_file_name:  
                self._postcards.append(line)
                
            self.parsePostcards()

    def parsePostcards(self):
        dates = []
        senders = []
        receivers = []
        
        for postcard in self._postcards:
            dates.append((postcard.split(";")[0]).split(":")[1])
            senders.append((postcard.split(";")[1]).split(":")[1])
            receivers.append((postcard.split(";")[2]).split(":")[1])
                
        for i, date in enumerate(dates, 1):
                if date not in self._date.keys():
                    self._date[date] = [i]
                else:
                    self._date[date].append(i)
                    
        for i, sender in enumerate(senders, 1):
                if sender not in self._from.keys():
                    self._from[sender] = [i]
                else:
                    self._from[sender].append(i)
                    
        for i, receiver in enumerate(receivers, 1):
                if receiver not in self._to.keys():
                    self._to[receiver] = [i]
                else:
                    self._to[receiver].append(i)
     
    def getNumberOfPostcards(self):
        return len(self._postcards)
    
    def getPostcardsByDateRange(self, date_range): 
        min_range = data_range[0]
        max_range = data_range[1]
        
        postcard_inrange = []
        for postcard in self._postcards:
            date = (postcard.split(";")[0]).split(":")[1]
            tmp = datetime.datetime.strptime(date, "%Y-%m-%d")
            if tmp >= min_range and tmp <= max_range:
                postcard_inrange.append(postcard)
            
        return postcard_inrange

    def getPostcardsBySender(self, sender):
        postcards_from_sender = []
        for postcard in self._postcards: 
            if sender == (postcard.split(";")[1]).split(":")[1]:
                postcards_from_sender.append(postcard)
        return postcards_from_sender
    
    def getPostcardsByReceiver(self, receiver):
        return [postcard for postcard in self._postcards if receiver == (postcard.split(";")[2]).split(":")[1]]

mypostcard = PostcardList()

mypostcard.readFile("exam_postcard_list0.txt")

print(mypostcard._date)
print(mypostcard._from)
print(mypostcard._to)

mypostcard.writeFile("exam_prova.txt")

mypostcard2 = PostcardList()

mypostcard2.readFile("exam_prova.txt")

new_postcards = ["date:2014-07-21; from:Daisy; to:Donald;", 
                 "date:2012-11-14; from:Goofy; to:Louie;"]

print(type(new_postcards))

mypostcard2.updateFile(new_postcards)

print(mypostcard2._postcards)
print(mypostcard2._date)
print(mypostcard2._from)
print(mypostcard2._to)

data_range = (datetime.datetime.strptime('2008-1-1', "%Y-%m-%d"), datetime.datetime.strptime('2009-06-09', "%Y-%m-%d"))

mypostcard2.getPostcardsByDateRange(data_range)

mypostcard2.getPostcardsBySender("Dewey")
mypostcard2.getPostcardsBySender("Mickey")

print("CIAO")

mypostcard2.getPostcardsByReceiver("Pluto")
mypostcard2.getPostcardsByReceiver("Happy")
```
